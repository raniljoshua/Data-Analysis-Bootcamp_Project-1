```python
#dependencies
#import requests
import pandas as pd
from census import Census
```

```python
#csv files
car_types_csv= "2023.csv"
car_prices_csv = "prices_clean.csv"

car_types_df= pd.read_csv(car_types_csv)
car_prices_df = pd.read_csv(car_prices_csv)

# Combine the data into a single dataset.  
car_data_complete = pd.merge(car_types_df, car_prices_df, how="left", on=["make", "model"])
car_data_complete.head()
```

```python
#drop any missing price data
car_data_complete.dropna(subset=['price'], inplace= True)
car_data_complete                         
```

```python
#delete extra year column
car_data_complete= car_data_complete .drop(columns=['year_y'])
car_data_complete
```

```python
car_data_complete_sorted = car_data_complete.sort_values(by=["body_styles", "price"])
car_data_complete_sorted
```
