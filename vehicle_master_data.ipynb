```python
import pandas as pd
from pathlib import Path
```

```python
# store the file paths into variables
vehicle_master_data = Path("vehicles.csv")
vehicle_price_data = Path('prices_clean.csv')
```

```python
# read in the files
vehicle_master_data_df = pd.read_csv(vehicle_master_data)
vehicle_price_data_df = pd.read_csv(vehicle_price_data)
```

```python
# remove the unnecessary columns then filter to 2023

vehicle_2023_master_df = vehicle_master_data_df[['year','make','model','comb08', 'cylinders', 'drive', 'fuelType', 'VClass', 'baseModel' ]]

vehicle_2023_master_df = vehicle_2023_master_df[vehicle_2023_master_df['year'] == 2023]
vehicle_2023_master_df
```

```python
# attempt to combine the two DFs
combined_vehicle_data = pd.merge(vehicle_price_data_df, vehicle_2023_master_df, on=['make', 'model'], how='left')
combined_vehicle_data = combined_vehicle_data[['year_x', 'make', 'model', 'price', 'VClass']]
combined_vehicle_data
```

```python
combined_vehicle_data.to_csv('combined_vehicle_data.csv', index=False)
```
